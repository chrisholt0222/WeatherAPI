```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Added to simplify weather data pull
import openweathermapy as ow

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# To convert Unix Date to readible date
from datetime import datetime

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Set variables for api data pull
i=0
wthr_data=[]
settings = {"units": "imperial", "appid": api_key}
views = {"summary": ["name","sys.country","dt","clouds.all","main.humidity","coord.lat","coord.lon",
                     "main.temp_max", "wind.speed"] }

# Begin data pull
print("Beginning Data Retrieval")
print("-------------------------------")

#Runs through the list of cities to retrieve data
for city in cities:
    i=i+1
    # Some cities may not be in the data, check to see if the can be pulled
    try:
        city_data = ow.get_current(city, **settings)
        wthr_data.append(city_data(*views["summary"]))
        print("Processing Record " + str(i) + " of Set " + " | " + city)
    except:
        print("City Not Found. Skipping ...")
    
print("-------------------------------")
print("Data Retrieval Complete")
print("-------------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# convert to datframe and add column labels
wthr_pd = pd.DataFrame(wthr_data)
columns = ["City","Country","Date","Cloudiness","Humidity","Lat","Lng","Max Temp","Wind Speed"]
wthr_pd.columns = columns

# covert date field to readible "YYYY-MM-DD time"
wthr_pd["Date_Read"] = pd.to_datetime(wthr_pd["Date"],unit="s")

# Send weather data to output file
wthr_pd.to_csv(output_data_file,header=True)

# Print record counts by field and first 5 records
print(wthr_pd.count())
print("")
wthr_pd.head()
```
